```python
#Charts
#orders based on pizza category

orders=df['pizza_category'].value_counts()
print(orders)
x=orders.index
print(x)
y=orders.values
print(y)

plt.bar(x,y,color='g')
plt.title("Orders based on pizza category")
plt.xlabel("Pizza category")
plt.ylabel("Orders")
plt.show()
```

```python
#Pie chart
#Orders based on pizza sizes

orders=df['pizza_size'].value_counts()
print(orders)
x=orders.values
lb=orders.index
e=[0.1,0,0,0,0]
k=['purple','pink','skyblue','yellow','brown']
print(lb)
plt.title("Orders based on pizza sizes",color='g')
plt.pie(x,labels=lb,autopct='%1.f%%',explode=e,colors=k)
plt.legend(lb,loc='upper left')
plt.show()
```

```python
#distribution of total price 
#scatter plot

price=df['total_price'].value_counts()
print(price)
x=price.index
y=price.values
plt.title("distribution of total price")
plt.xlabel("Total price")
plt.ylabel("Counts")
plt.scatter(x,y,color='g')
plt.show()
```

```python
#distribution of total price
#line chart

y=df.loc[300:350,'total_price']
print(y)
plt.plot(y,'gs--')
plt.title("Distribution of total price",color='purple')
plt.xlabel("Counts",color='purple')
plt.ylabel("Total price",color='purple')
plt.show()
```

```python
#orders on pizza size whose price is greater than 30

orders=df[df['total_price']>30]['pizza_size']
print(orders)
a=orders.value_counts()
print(a)
x=a.index
y=a.values
plt.bar(x,y,color='orange')
plt.show()
```

```python
#orders on pizza category where total price is less than 40

orders=df[df['total_price']<40]['pizza_category']
print(orders)
a=orders.value_counts()
print(a)
x=a.index
y=a.values
plt.bar(x,y,color='skyblue')
plt.show()
```

```python
#distribution of unit price
#histogram

price=df['unit_price']
plt.title("distribution of unit price")
plt.xlabel("Unit price")
plt.ylabel("counts")
plt.hist(price,color='purple')
plt.show()
```

```python
#top 10 best selling pizzas
#popular=df.groupby('pizza_name')['quantity'].sum().sort_values().tail(10)
popular=df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False).head(10)
print(popular)
popular.plot(kind='bar')
plt.title("Top 10 best selling pizzas")
plt.show()
```

```python
#method 2
popular=df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False).head(10)
print(popular)
x=popular.index
y=popular.values
plt.barh(x,y,color='g')
plt.title("Top 10 best selling pizzas")
plt.xlabel("Sales")
plt.ylabel("Pizza Name")
plt.show()
```
